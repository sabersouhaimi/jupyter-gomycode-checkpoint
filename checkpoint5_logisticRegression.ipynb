```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_titanic.csv",sep = ";")
df
```

```python
df.isna().sum()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace = True)
df
```

```python
df.drop('Cabin',inplace = True,axis = 1)
```

```python


Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                    "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                    "Lady" :      "Royalty",

                    "the Countess" : "Royalty",
    
                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"}
```

```python
q = []

for i in df["Name"]:
    title = i.split(".")[0].split(",")[1]
    q.append(title)
        
df["Status"] = q
```

```python
df.drop("Name",axis=1,inplace = True)
df
```

```python
df.info()
```

```python
df["Sex"] = df["Sex"].map({"male":1,"female":0})
df.drop("PassengerId",inplace = True,axis = 1)
df["Survived"] = df["Survived"].map({'No':0,"Yes":1})
df.drop("Ticket",inplace = True,axis = 1)
```

```python
df
```

```python
df["Embarked"].unique()
```

```python
df["Embarked"] = df["Embarked"].map({"S":0,"C":1,"Q":2})
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

df['Status'] = lb.fit_transform(df["Status"])
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df["Survived"] = df["Survived"].astype(int)
df["Survived"].unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

x = df[df.columns[1:]].values
y = df["Survived"].values.reshape(-1,1)

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=0)
x.shape,y.shape
```

```python
lg = LogisticRegression()
lg.fit(xtrain,ytrain)
y_pred = lg.predict(xtest)
print(lg.score(xtest,ytest))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred))
```

ROC:

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(ytest, y_pred)

print("False Positive Rates:", fpr)
print("True Positive Rates:", tpr)
print("Thresholds:", thresholds)
```

AUC:

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(ytest, y_pred)

# calculate the AUC score
roc_auc = auc(fpr, tpr)
print(roc_auc)
```
